```python
import pandas as pd
import sqlite3
from pathlib import Path
```

```python
csv_file_path = '../csv-data/wineQualityReds.csv'

df = pd.read_csv(csv_file_path, index_col=0)
df['id'] = range(1, len(df) + 1)

df
```

```python
csv_file_path2 = '../csv-data/predictionModelList.csv'

df2 = pd.read_csv(csv_file_path2)

df2
```

```python
csv_file_path3 = '../csv-data/nullDataSolutions.csv'

df3 = pd.read_csv(csv_file_path3)

df3
```

```python
sqlite_db_path = Path('../database/basic-forecast.db')
db_dir = sqlite_db_path.parent
db_dir.mkdir(parents=True, exist_ok=True)
```

```python
conn = sqlite3.connect(sqlite_db_path)
```

```python
df.to_sql('red_wine_quality', conn, if_exists='replace', index=False)
df2.to_sql('prediction_model', conn, if_exists='replace', index=False)
df3.to_sql('null_data_solutions', conn, if_exists='replace', index=False)
```

```python
# CHECK DATA
query = "SELECT * FROM red_wine_quality"
df_wine = pd.read_sql_query(query, conn)

query = "SELECT * FROM prediction_model"
df_prediction = pd.read_sql_query(query, conn)

query = "SELECT * FROM null_data_solutions"
df_null_sol = pd.read_sql_query(query, conn)

conn.close()
```
